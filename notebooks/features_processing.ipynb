```python
parole = []
for i in range(len(df)):
    parole.append(df.loc[i,'features'])
```

```python
parole_1 = []
for elem in parole:
    parole_1.append(elem.split(" "))
```

```python
parole = []
for l in parole_1:
    for parola in l:
        parole.append(parola)
```

```python
parole = ['balcone', 'terrazza', 'cantina', 'piscina', 'giardino']
```

```python
for parola in parole:
    df[parola] = 0
```

```python
df['features'] = df['features'].str.lower()
df['features']
```

```python
for parola in parole:
    if parola == 'balcone':
        df.loc[(df['features'].str.contains(parola)) | df['features'].str.contains('balconi'), parola] = 1
    else:
         df.loc[(df['features'].str.contains(parola)), parola] = 1
```
